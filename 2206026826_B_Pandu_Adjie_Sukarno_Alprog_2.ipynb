```python
# Fungsi cekNama digunakan untuk ngecek tiap string yang diinput saat memasukkan nama mahasiswa
def cekNama(string):
  for string in string: # Melakukan iterasi untuk tiap elemen pada string
    if string == '0' or string == '1' or string == '2' or string == '3' or string == '4' or string == '5' or string == '6' or string == '7' or string == '8' or string == '9':
          return False # Fungsi akan mengembalikan nilai false jika pada input string terdapat angka
```

```python
# Fungsi inputDataBaru merupakan fungsi yang dipanggil ketika user ingin melakukan input data mahasiswa

def inputDataBaru():
  print("-------------------------------------------------------")
  print("                    Input Data Baru                    ")
  print("-------------------------------------------------------")

  dataBaruMahasiswa = [] # Mendefinisikan variabel dataBaruMahasiswa sebagai list
  nama = ""     # Mendefinisikan variabel nama
  npm = ""      # Mendefinisikan variabel npm
  jurusan = ""  # Mendefinisikan variabel jurusan
  kelas = ""    # Mendefinisikan variabel kelas

  while True:   # Gunakan while loop yang selalu melakukan pengulangan
    nama = input("Nama : ") # input nama kedalam variabel "nama"
    if cekNama(nama) == False: # melakukan pengecekan. Jika fungsi cekNama dengan parameter "nama" mengeluarkan output False,
      print("Nama hanya berisi huruf") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
    else: # Jika fungsi cekNama mengeluarkan output True,
      break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

  while True:  # Gunakan while loop yang selalu melakukan pengulangan
    npm = input("NPM : ") # input npm kedalam variabel "npm"
    if npm.isnumeric() == False: # Melakukan pengecekan. Jika npm yang diinput mengandung karakter yang bukan angka,
      print("NPM hanya berisi angka.") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
    else: # Jika npm yang diinput, sudah merupakan angka semua,
      break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

  while True:  # Gunakan while loop yang selalu melakukan pengulangan
    jurusan = input("Jurusan [Matematika/Statistika/Aktuaria] : ") # Input jurusan kedalam variabel "jurusan"
    if jurusan != "Matematika" and jurusan != "Statistika" and jurusan != "Aktuaria": # Melakukan pengecekan. Jika jurusan yang diinput bukan Matematika, Statistika, atau Aktuaria,
      print("Jurusan hanya antara Matematika, Statistika, atau Aktuaria") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
    else: # Jika jurusan yang diinput sudah sesuai yang diminta,
      break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

  while True:  # Gunakan while loop yang selalu melakukan pengulangan
    kelas = input("Kelas Algoritma dan Pemrograman [A/B/C/D] : ") # Input kelas kedalam variabel "kelas"
    if kelas != "A" and kelas != "B" and kelas != "C" and kelas != "D": # Melakukan pengecekan. Jika kelas yang diinput bukan A, B, C, atau D,
      print("Kelas hanya antara A, B, C, atau D") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
    else: # Jika kelas yang diinput sudah sesuai yang diminta,
      break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.


  dataBaruMahasiswa.append(nama) # Memasukkan nilai dari variabel "nama" kedalam list dataBaruMahasiswa
  dataBaruMahasiswa.append(npm)  # Memasukkan nilai dari variabel "npm" kedalam list dataBaruMahasiswa
  dataBaruMahasiswa.append(jurusan) # Memasukkan nilai dari variabel "jurusan" kedalam list dataBaruMahasiswa
  dataBaruMahasiswa.append(kelas)  # Memasukkan nilai dari variabel "kelas" kedalam list dataBaruMahasiswa

  databaseMahasiswa.append(dataBaruMahasiswa) # Memasukkan nilai dari list dataBaruMahasiswa kedalam list databaseMahasiswa

  print("              Data berhasil ditambahkan!\n             ")
```

```python
# Fungsi modifikasiData merupakan fungsi yang dipanggil ketika user ingin melakukan perubahan atau koreksi dari data yang telah diinput sebelumnya.
def modifikasiData():
  print("-------------------------------------------------------")
  print("                   Modifikasi Data                     ")
  print("-------------------------------------------------------")

  editID = int(input("Input ID Mahasiswa : ")) # Melakukan input ID mahasiswa untuk menentukan data mana yang akan dilakukan perubahan
  dataModifikasi = [] # Mendefinisikan variabel dataModifikasi sebagai list
  nama2 = ""   # Mendefinisikan variabel nama2
  npm2 = "" # Mendefinisikan variabel npm2
  jurusan2 = "" # Mendefinisikan variabel jurusan2
  kelas2 = "" # Mendefinisikan variabel kelas2

  while True:
    if editID > len(databaseMahasiswa) - 1: # Jika ID yang diinput tidak ada,
      print("ID tidak tersedia") # Maka akan di print kalimat disamping
      break  # Dan pemanggilan fungsi dihentikan
    else:

    # Lakukan proses yang sama seperti saat menginput data di fungsi inputDataBaru
      while True:  # Gunakan while loop yang selalu melakukan pengulangan
        nama2 = input("Nama : ") # input nama kedalam variabel "nama2"
        if cekNama(nama2) == False: # melakukan pengecekan. Jika fungsi cekNama dengan parameter "nama2" mengeluarkan output False,
          print("Nama hanya berisi huruf") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
        else:  # Jika fungsi cekNama mengeluarkan output True,
          break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

      while True: # Gunakan while loop yang selalu melakukan pengulangan
        npm2 = input("NPM : ") # input npm kedalam variabel "npm2"
        if npm2.isnumeric() == False: # Melakukan pengecekan. Jika npm yang diinput mengandung karakter yang bukan angka,
          print("NPM hanya berisi angka.") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
        else: # Jika npm yang diinput, sudah merupakan angka semua,
          break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

      while True: # Gunakan while loop yang selalu melakukan pengulangan
        jurusan2 = input("Jurusan [Matematika/Statistika/Aktuaria] : ") # Input jurusan kedalam variabel "jurusan2"
        if jurusan2 != "Matematika" and jurusan2 != "Statistika" and jurusan2 != "Aktuaria": # Melakukan pengecekan. Jika jurusan yang diinput bukan Matematika, Statistika, atau Aktuaria,
          print("Jurusan hanya antara Matematika, Statistika, atau Aktuaria") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
        else: # Jika jurusan yang diinput sudah sesuai yang diminta,
          break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

      while True: # Gunakan while loop yang selalu melakukan pengulangan
        kelas2 = input("Kelas Algoritma dan Pemrograman [A/B/C/D] : ") # Input kelas kedalam variabel "kelas2"
        if kelas2 != "A" and kelas2 != "B" and kelas2 != "C" and kelas2 != "D": # Melakukan pengecekan. Jika kelas yang diinput bukan A, B, C, atau D,
          print("Kelas hanya antara A, B, C, atau D") # Maka akan diprint tulisan disamping dan while loop terus berlanjut.
        else: # Jika kelas yang diinput sudah sesuai yang diminta,
          break # maka while loop akan dihentikan dan lanjut ke proses selanjutnya.

      dataModifikasi.append(nama2) # Memasukkan nilai dari variabel "nama2" kedalam list dataModifikasi
      dataModifikasi.append(npm2)  # Memasukkan nilai dari variabel "npm2" kedalam list dataModifikasi
      dataModifikasi.append(jurusan2) # Memasukkan nilai dari variabel "jurusan2" kedalam list dataModifikasi
      dataModifikasi.append(kelas2) # Memasukkan nilai dari variabel "kelas2" kedalam list dataModifikasi

      databaseMahasiswa[editID] = dataModifikasi # Memasukkan nilai dari list dataModifikasi kedalam list databaseMahasiswa dengan ID sesuai yang diinput.


      print("              Data berhasil dimodifikasi.\n            ")
      break
```

```python
# Fungsi hapusData merupakan fungsi yang dipanggil ketika user ingin menghapus data yang telah diinput sebelumnya
def hapusData():
  print("-------------------------------------------------------")
  print("                      Hapus Data                       ")
  print("-------------------------------------------------------")

  deleteID = int(input("Input ID Mahasiswa : ")) # Memasukkan ID dari data yang ingin dihapus
  while True:
    if deleteID > len(databaseMahasiswa) - 1: # Jika ID yang diinput tidak ada,
      print("ID tidak tersedia") # Maka akan di print kalimat disamping
      break  # Dan pemanggilan fungsi dihentikan
    else:
      del databaseMahasiswa[deleteID] # Menghapus data pada databaseMahasiswa sesuai dengan ID yang telah diinput dengan del.
      print("                Data berhasil dihapus.\n               ")
      break
```
